## Условие 3: Задача 3
- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True, linewidths=.5, cmap='coolwarm');
```

**Вывод:** На данной тепловой карте видна зависиомость представленных значений. Так, к примеру, из диаграммы мы видим, что видовые характеристики дома коррелировают с ценой на 40%, с жилой квадратурой на 30%  и с наличием набережной 40%. 

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=5)

plt.xlabel('Цена')
plt.ylabel('Оценка')
plt.title('Распределения цены домов в завистимости от их оценки ');
```

**Вывод:** Выше представлен боксплот визуализации распределения цены домов в завистимости от их оценки. Линия посередине ящиков - это медиана переменной price ( ось Х - цена). Меньшая часть наблюдений имеют значение выше этой медианы, а большая - ниже ее. Разница между средним значением и медианой указывает на то, что существует несколько наблюдений с экстремальными значениями, которые увеличивают среднее значение. 

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

**Вывод:** Данный график показывает нормальное распределение цены дома и его жилой квадратуры. Как видим, чем больше квадратура помещения, тем выше цена на него. Наибольшее количество занимают дома с наименьшей квадратурой и наименьшей стоимостью. В то же время, имеются отклонения от нормы распределения. 

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

**Вывод:**  Данный график показывает нормальное распределение цены дома и его грейда(оценки). Как видим, чем выше грейд помещения, тем выше цена на него. Больше всего в придложении домов со средним грейдом.В то же время, имеются отклонения от нормы распределения.

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=5)

plt.xlabel('Цена')
plt.ylabel('Набережная')
plt.title('Распределение цены дома в зависимости от наличия набережной');
```

**Вывод:** Представлен боксплот визуализации распределения цены дома в зависимости от наличия набережной. Линия посередине ящиков - это медиана переменной price ( ось Х - цена).Половина наблюдений имеют значение выше этой медианы, а половина - ниже ее. Как и среднее значение, медиана является мерой положения центра распределения. В отличие от среднего значения, наблюдения с экстремальными значениями оказывают на нее меньшее влияние. Как видим на графике, при наличии набережной растет цена дома.
